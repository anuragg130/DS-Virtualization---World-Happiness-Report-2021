```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns
```

Reading DF using pandas inbuilt function

```python
df = pd.read_csv('world-happiness-report-2021.csv')
```

```python
df.head()
```

Check for Null Values in the Dataframe

```python
df.isnull().sum()
```

Check for duplicated values

```python
df.duplicated().sum()
```

```python
df.columns
```

Figuring out all of the useful columns from the dataframe

```python
cols=['Country name', 'Regional indicator', 'Ladder score','Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
```

```python
df = df[cols]
```

```python
df.rename({'Country name':'country','Regional indicator':'region',
          'Logged GDP per capita':'GDP_per_capita','Social support':'social_score',
          'Healthy life expectancy':'life_expectancy',
           'Freedom to make life choices':'freedom_choice','Generosity':'generosity',
          'Perceptions of corruption':'corruption_score',
          'Ladder score':'happiness_score'},axis=1,inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure()
sns.scatterplot(x='GDP_per_capita',y='happiness_score',hue='region',data=df)
plt.legend(prop={'size':5.9})
plt.title('Relation between Happiness and GDP')
```

```python
plt.figure()
sns.scatterplot(x='freedom_choice',y='happiness_score',hue='region',data=df)
plt.legend(prop={'size':5.9})
plt.title('Relation between Freedom of Choice and Happiness Score')
```

```python
happiness_region=df.groupby('region').mean()
```

```python
happiness_region.head()
```

```python
plt.figure()
sns.scatterplot(x='GDP_per_capita',y='happiness_score',hue='region',data=happiness_region)
plt.legend(prop={'size':5.9})
plt.title('Region wise GDP and happiness')
```
